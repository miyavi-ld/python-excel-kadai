```python
import pandas as pd
import openpyxl
```

```python
file_list = ['2022_年間売上表.xlsx', '2023_年間売上表.xlsx']

df = pd.DataFrame()
for file_name in file_list:
    temp_df = pd.read_excel(file_name)
    temp_df.columns = ['売上年', '商品', '金額（千円）']

    df = pd.concat([df, temp_df], ignore_index=True)
df_summary = df.groupby(['商品', '売上年'], as_index=False)['金額（千円）'].sum()
print(df_summary)
```

```python
df_summary.to_excel('売上集計表.xlsx', sheet_name='Sheet1', index=False)
```

```python
from openpyxl.styles import PatternFill
wb = openpyxl.load_workbook('売上集計表.xlsx')
ws = wb['Sheet1']

header_fill = PatternFill(start_color='F2F2F2', end_color='F2F2F2', fill_type='solid')
for cell in ws[1]: # ws[1] は1行目を意味します
    cell.fill = header_fill

wb.save('売上集計表.xlsx')
```
