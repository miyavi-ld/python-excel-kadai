```python
import pandas as pd

df = pd.DataFrame({
    '日付': ['2023-05-17', '2023-05-18', '2023-05-19', '2023-05-20', '2023-05-21'],
    '社員名': ['山田' ,'佐藤' ,'鈴木' ,'田中' ,'高橋'],
    '売上': [100 ,200, 150 ,300 ,250],
    '部門': ['メーカー' ,'代理店' ,'メーカー' ,'商社' ,'代理店'],
})
```

```python
df['平均売上'] = df['売上'].mean()
```

```python
def performance(individual_sales):
  # 全体の平均売上をスカラー値として取得
  # df['平均売上']列はすでに全体の平均値(200.0)で埋められているため、iloc[0]でその値を取得する
  overall_average_sales = df['平均売上'].iloc[0]

  if individual_sales >= overall_average_sales + 50:
    return 'A'
  elif individual_sales >= overall_average_sales:
    return 'B'
  else:
    return 'C'

df['業績ランク'] = df['売上'].apply(performance)
```

```python
writer = pd.ExcelWriter('業績.xlsx')
df.to_excel(writer, sheet_name='Sheet1', index=False)
writer.close()
```
